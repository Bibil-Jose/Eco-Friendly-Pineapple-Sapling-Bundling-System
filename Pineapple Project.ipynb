# Pineapple Sapling Winding Machine Simulation

```julia
using Plots
using Random

function simulate_stochastic_winding_machine(num_revolutions, adhesive_length, container_radius)
    # Initialize variables
    roller_position = 0
    bundle_tightness = 0
    strengths_needed = Float64[]

    # Lists to store data for plotting
    time_points = []
    roller_positions = []
    bundle_tightness_values = []

    # Set a random seed for reproducibility
    Random.seed!(123)

    # Simulation loop
    for revolution in 1:num_revolutions
        # Simulate stochastic strength needed
        strength_needed = rand(10:20)  # Adjust the range as needed
        push!(strengths_needed, strength_needed)

        # Oscillatory movement of the cylindrical container
        oscillation_angle = sin(revolution / num_revolutions * 2 * π)
        container_height = oscillation_angle * container_radius

        # Rotary movement of roll casing and driving wheel
        roller_position += adhesive_length
        bundle_tightness = (revolution % 2) * adhesive_length / 2

        # Record simulation data
        push!(time_points, revolution)
        push!(roller_positions, roller_position)
        push!(bundle_tightness_values, bundle_tightness)
    end

    # Plot the simulation results
    plot(time_points, roller_positions, label="Roller Position", xlabel="Time (Revolutions)", ylabel="Position")
    plot!(time_points, bundle_tightness_values, label="Bundle Tightness")

    # Scatter plot for stochastic strength needed
    scatter!(time_points, strengths_needed, label="Strength Needed", markersize=5)

    # Plot distribution of strength needed
    histogram(strengths_needed, bins=10, label="Strength Distribution", xlabel="Strength Needed", ylabel="Frequency")

    # Plot impact of strength variations on bundle tightness
    plot(time_points, strengths_needed .+ bundle_tightness_values, label="Strength + Bundle Tightness", xlabel="Time (Revolutions)", ylabel="Combined Effect")

    title!("Stochastic Simulation of Pineapple Sapling Winding Machine")
    plot!()  # This line adds the legend
end

# Set simulation parameters
num_revolutions = 10
adhesive_length = 5  # cm
container_radius = 10  # cm

# Run the stochastic simulation
simulate_stochastic_winding_machine(num_revolutions, adhesive_length, container_radius)
```

# Satistical Analysis

```julia
using Plots
using Statistics  # Import the Statistics module for mean and std functions

# Example data
time_points = 1:10
bundle_tightness_values = rand(10)

# Statistical analysis
mean_values = [mean(bundle_tightness_values) for _ in time_points]
std_dev_values = [std(bundle_tightness_values) for _ in time_points]

# Plot
plot(time_points, bundle_tightness_values, label="Bundle Tightness")
plot!(time_points, mean_values, label="Mean Bundle Tightness")
plot!(time_points, std_dev_values, label="Standard Deviation of Bundle Tightness")
title!("Statistical Analysis of Bundle Tightness Over Time")
savefig("statistical_analysis.png")
```
